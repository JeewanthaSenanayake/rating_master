```python
df_title_with_data = pd.merge(df_rating_and_title,df_title_principles_unic, on='tconst',how='inner')
df_title_with_data.head()
```

```python
columns_to_drop = ['archive_footage','archive_sound','composer','production_designer','self','producer','numVotes']
df_title_with_data = df_title_with_data.drop(columns=columns_to_drop)
df_title_with_data.head()
```

```python
df_title_with_data.replace('\\N', 'Unknown', inplace=True)
```

```python
df_title_with_data.head()
```

```python
import numpy as np
```

```python
def combine_columns(row):
    if(row['actor'] == 'Unknown' and row['actress'] == 'Unknown'):
        return "Unknown"
    elif(row['actor'] == 'Unknown'):
        return row['actress']
    elif(row['actress']== 'Unknown'):
        return row['actor']
    else:
        return row['actor'] + row['actress']
    
df_title_with_data['actor_actress'] = df_title_with_data.apply(combine_columns, axis=1)
```

```python
df_title_with_data.head()
```

```python
columns_to_drop = ['actor','actress','cinematographer']
df_title_with_data = df_title_with_data.drop(columns=columns_to_drop)
df_title_with_data.head()
```

```python
# Create a dictionary to map codes to names
code_to_name = dict(zip(df_persons_name['nconst'], df_persons_name['primaryName']))

# Define a function to replace codes with names using NumPy
def replace_codes_with_names(cell):
    codes = cell.split(',')
    names = np.array([code_to_name.get(code, 'Unknown') for code in codes], dtype=object)
    return ','.join(names)
#     return names

# Replace director codes with names
df_title_with_data['director'] = df_title_with_data['director'].apply(replace_codes_with_names)
df_title_with_data['editor'] = df_title_with_data['editor'].apply(replace_codes_with_names)
df_title_with_data['writer'] = df_title_with_data['writer'].apply(replace_codes_with_names)
df_title_with_data['actor_actress'] = df_title_with_data['actor_actress'].apply(replace_codes_with_names)
df_title_with_data.head()
```
