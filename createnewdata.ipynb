```python
import pandas as pd
```

```python
df_title_principals = pd.read_csv("/kaggle/input/imdbdatasets/title_principals.tsv",sep='\t')
```

```python
# Pivot the DataFrame to create the new table
table = df_title_principals.pivot_table(index='tconst', columns='category', values='nconst', aggfunc=lambda x: '\\N' if x.isnull().all() else ','.join(x))

# Reset the index to have 'tconst' as a column instead of an index
table.reset_index(inplace=True)

# Fill any remaining NaN values with '\N'
table = table.fillna('\\N')

#Download
tsv_data = table.to_csv(sep='\t', index=False)

# Specify the file path within the Kaggle environment
file_path = '/kaggle/working/title_principals_unic.tsv'

# Write the TSV data to the file within the Kaggle environment
with open(file_path, 'w') as file:
    file.write(tsv_data)
    
table.head()
```
