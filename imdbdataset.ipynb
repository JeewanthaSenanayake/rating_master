# **Merge**

```python
df_rating_akas = pd.merge(df_title_ratings, df_title_akas , left_on='tconst', right_on='titleId', how='inner')
```

```python
df_rating_akas.head()
```

```python
# Remove rows with NaN or '\N' values specifically from columns
column_names = ['language']
for column_name in column_names:
    df_rating_akas[column_name] = df_rating_akas[column_name].replace({'\\N': None})
    
df_rating_akas = df_rating_akas.dropna(subset=column_names) 
df_rating_akas.head()
```

```python
df_rating_akas.shape
```

```python
search_term =  r'[a-zA-Z]'
mask = df_rating_akas['title'].str.contains(search_term, case=False, na=False)
result = df_rating_akas[mask]

columns_to_drop = ['types','attributes']
df_rating_akas = result.drop(columns=columns_to_drop)
```

```python
df_rating_akas.head()
```

```python
df_rating_akas.shape
```

```python
result1 = pd.merge(df_rating_akas, df_title_basics_new , on='tconst', how='inner')

columns_to_drop = ['title','titleId','isOriginalTitle','ordering']
result2 = result1.drop(columns=columns_to_drop)

# Specify the column to check for repeated values
column_name = 'tconst'

# Create a boolean mask to identify rows where column has the same value as the previous row
mask = result2[column_name].duplicated(keep='first')

# Invert the mask to keep rows with unique values in column
df_rating_and_titles = result2[~mask]
```

```python
df_rating_and_titles.head()
```

```python
columns_to_drop = ['region','originalTitle']
df_rating_and_titles = df_rating_and_titles.drop(columns=columns_to_drop)
df_rating_and_titles.head()
```
